```python
from bs4 import BeautifulSoup
import requests
import random
```

```python

url="https://www.goodreads.com/quotes/tag/inspirational"


response=requests.get(url).text
# response=response.content
```

```python
soup=BeautifulSoup(response,'lxml')
# print(soup)
soup
```

```python
match=soup.find('div', class_='quoteText')
texts=match.contents[0].strip()   #get a sample quote 
print(texts)
```

```python
print(match.span.text.strip())    #print a sample author
```

```python
match=soup.findAll('div', class_='quoteText')     #match is a list contains all divs of class quote texts
type(match)
# author=match[-1].span.text.strip()
# quote=match[0].contents[0].strip()
match[0]
```

```python
quote_list=[]
for i in match:
    # print(i.contents[0].strip())
    author=i.span.text.strip()
    quote=i.contents[0].strip()
    message={'quote':quote,'author':author}
    quote_list.append(message)
    
    
random.choice(quote_list)
    
```

```python
def random_quotes():
    import random
    choosed=random.choice(quote_list)
    print(f"{choosed['quote']} \nby\n{choosed['author']} ")
```

```python
random_quotes()
```
